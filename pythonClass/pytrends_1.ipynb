```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import pandas
import numpy
import time

# Login to Google. Only need to run this once, the rest of requests will use the same session.
pytrend = TrendReq(hl='en-US', tz=360)

kw_list=["神力女超人1984", "Adidas", "Asics"]

pytrend.build_payload(kw_list, cat=0, timeframe='today 3-m',geo='TW',gprop='')

# Interest Over Time
interest_over_time_df = pytrend.interest_over_time()
#time.sleep(randint(5, 10))

print(interest_over_time_df)
avg_kw = interest_over_time_df.mean()
print(f"The average interest in the last 3 months is\n", avg_kw, sep="")

plt.plot (interest_over_time_df.iloc[:,0], label = kw_list[0])
plt.plot (interest_over_time_df.iloc[:,1], label = kw_list[1])
plt.plot (interest_over_time_df.iloc[:,2], label = kw_list[2])

plt.xticks(rotation=45)
plt.xlabel('Time')
plt.ylabel('Level of Interest')
plt.title('Interest over time')
plt.legend()
plt.show()
```
